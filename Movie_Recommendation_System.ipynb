```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Load dataset
movies = pd.read_csv("movies.csv")

# Extract movie features
features = movies.drop("title", axis=1)

# Calculate similarity matrix
similarity = cosine_similarity(features)

# Create movie index mapping
movie_index = pd.Series(movies.index, index=movies['title'])

# Recommendation function
def recommend(movie_name):

    if movie_name not in movie_index:
        print("Movie not found in dataset")
        return

    idx = movie_index[movie_name]

    similarity_scores = list(enumerate(similarity[idx]))

    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    print("\nRecommended Movies:\n")

    for i in similarity_scores[1:6]:
        print(movies['title'][i[0]])

# User input
movie = input("Enter movie name: ")

recommend(movie)
```
